```python
import pickle as pkl
with open("latest_overall_results.pkl","rb") as f: 
    res = pkl.load(f)
```

```python
games_won = {}
for episodes,  results in res.items():
    games_won[episodes] = results[0]
```

```python
import pandas as pd 
```

```python
res_df = pd.DataFrame(games_won.items(), columns=["episodes","games_won"])
```

```python
res_df.plot(x="episodes", y="games_won")
```
